```python
import os
import gym 
from stable_baselines3 import DQN
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
#environment_name = "Pong-v0"
#rewards = [[1,0,5], [1,10,4], [1,4,6]]
#rewards = [[1,0,8], [1,10,3]]
#rewards = [[1,7,5],[1,0,4]]
rewards = [[1,1,6],[1,3,4]]
```

```python
#env = gym.make(environment_name)
from GridEnv import WindyGridworld
env = WindyGridworld(
        height=20,
        width=11,
        rewards=rewards,
        wind=True,
        allowed_actions=['L', 'R', 'C'],
        reward_terminates_episode=False
    )
```

```python
log_path = os.path.join('Training', 'Logs')
load_path = os.path.join('Training', 'Saved Models','Env2_Exp0.1_2M')
model = DQN.load(load_path, env = env)
```

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True)
env.close()
```
